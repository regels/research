```python
import pandas as pd
import numpy as np

import nfl_data_py as nfl

from great_tables import GT

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Get play by play data
pbp = nfl.import_pbp_data([2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
pbp.head()
```

```python
pbp.shape
```

```python
pbp.head(20)
```

```python
# Filter for only run or pass plays
second_down = pbp.query('play_type == "run" | play_type == "pass"').reset_index()
second_down.shape
```

```python
# Filter plays that were 2nd down, distance to go between 8 and 12 yards, yardline between own 20 and opponent 20
second_down = second_down.query('down == 2.0 & ydstogo >= 8 & ydstogo <= 12 & yardline_100 > 20 & yardline_100 < 80').reset_index()
second_down.shape
```

```python
# Filter only for plays where the series result was a first down or punt. I do not want series of downs that end in turnovers, touchdowns etc included
second_down = second_down.query('series_result == "Punt" | series_result == "First down"')
second_down.shape
```

```python
second_down = second_down[['play_type', 'down', 'ydstogo', 'series_result']]
second_down.shape
```

```python
second_down.head(20)
```

```python
second_down['punt'] = np.where(second_down['series_result']=='Punt', 1, 0)
second_down['first down'] = np.where(second_down['series_result']=='First down', 1, 0)
second_down.head(20)
```

```python
grouped = second_down.groupby('play_type').agg({
    'play_type':'count',
    'punt':'sum',
    'first down':'sum'
})
grouped = grouped.rename(columns={'play_type': 'plays'})
grouped['punt %'] = (grouped['punt']/grouped['plays'])
grouped['first down %'] = (grouped['first down']/grouped['plays'])
grouped = grouped.rename(columns={'play_type': 'play type'})
grouped.reset_index(inplace=True)
grouped
```

```python
# Create a display table using great_tables
(
    GT(grouped)
    .tab_header(title='To Pass or Run on 2nd and Long', subtitle='NFL 2014-2023, Series result on 2nd down and 8-12 yds to go, between own 20 and opponent 20')
    .tab_stub(rowname_col='play_type')
    .fmt_number(columns='plays', decimals=0, compact=False)
    .fmt_number(columns='punt', decimals=0, compact=False)
    .fmt_number(columns='first down', decimals=0, compact=False)
    .fmt_percent(columns='punt %', decimals=2)
    .fmt_percent(columns='first down %', decimals=2)
)
```
